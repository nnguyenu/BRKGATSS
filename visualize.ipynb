```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
for i in range(10):
    G.add_node(i, time='5pm', color='red')
print(G.nodes[1])
nx.draw(G, with_labels=True, font_weight='bold', node_color=[G.nodes[i]['color'] for i in G.nodes])
```

```python
G = nx.Graph()
# node from 1 to 10, node color is black by default
for i in range(1, 11):
    G.add_node(i, color='green')
G.add_edges_from([(1, 2), (2, 3), (2, 4), (2, 7), (2, 8), (3, 4), (4, 5), (4, 6), (4, 7), (6, 9), (7, 8), (7, 9), (9, 10)])

G.nodes[6]['color'] = 'orange'
G.nodes[7]['color'] = 'orange'
pos = nx.spring_layout(G)
nx.draw(G, pos=pos, with_labels=True, node_color=[node[1]['color'] for node in G.nodes(data=True)])
plt.show()

G.nodes[8]['color'] = 'orange'
G.nodes[9]['color'] = 'orange'
nx.draw(G, pos=pos, with_labels=True, node_color=[node[1]['color'] for node in G.nodes(data=True)])
plt.show()

G.nodes[10]['color'] = 'orange'
nx.draw(G, pos=pos, with_labels=True, node_color=[node[1]['color'] for node in G.nodes(data=True)])
plt.show()
```
